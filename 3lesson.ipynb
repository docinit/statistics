```python
'''
1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150.
Посчитать (желательно без использования статистических методов наподобие std, var, mean)
среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.
'''

values = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]

mean = sum(values)/len(values)
std = (sum([(i - mean)**2 for i in values]) / (len(values)-1))**0.5
var_biased = sum([(i - mean)**2 for i in values]) / (len(values))
var = sum([(i - mean)**2 for i in values]) / (len(values)-1)

print('''Среднее = %.2f,
среднее квадратичное отклонение = %.2f,
смещенная дисперсия = %.2f,
несмещенная дисперсия = %.2f
'''%(mean, std, var_biased, var))
```

```python
'''
2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых.
Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?
'''
import numpy as np

first_box = 8
first_box_white = 5
second_box = 12
second_box_white = 5

#####Ровно 3#######
# 3 белые из второй коробки
second_var = np.math.comb(second_box_white,3) * np.math.comb(second_box-second_box_white,1)
first_var = np.math.comb(first_box-first_box_white,2)
total_var = second_var * first_var
# 1 белый из первой коробки, 2 белых из второй коробки
second_var = np.math.comb(second_box_white,2) * np.math.comb(second_box-second_box_white,2)
first_var = np.math.comb(first_box_white,1)*np.math.comb(first_box-first_box_white,1)
total_var += second_var * first_var
# 2 белый из первой коробки, 1 белых из второй коробки
second_var = np.math.comb(second_box_white,1) * np.math.comb(second_box-second_box_white,3)
first_var = np.math.comb(first_box_white,2)
total_var += second_var * first_var

print('Вероятность вытянуть ровно 3 белых шара = ',total_var / (np.math.comb(second_box,4) * np.math.comb(first_box,2)))
```

```python
'''
2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых.
Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?
'''

from itertools import product
#####Хотя бы 3#######
# содержимое первой коробки
first_box = 8
first_box_white = 5
first_box_black = first_box-first_box_white
# содержимое второй коробки
second_box = 12
second_box_white = 5
second_box_black = second_box-second_box_white
# количество выбранных шаров и кол-во шаров, которые должны быть белыми
numbers = 4
number_from_first_box = 2
number_from_second_box = 4
white_balls = 3 # минимальное кол-во белых шаров, которые надо вытянуть
# считаем вероятность в цикле;
total_var = 0
# с помощью модуля функции product  в модуле itertools переберем возможные комбинации кол-ва белых шаров из двух коробок
for first_white_balls,second_white_balls in product(range(number_from_first_box+1),range(number_from_second_box+1)):
    if first_white_balls + second_white_balls in range(white_balls,numbers+1):
        # если так случилось, что существует вариант, при котором из первой коробки может быть вытянуто 0 белых шаров,
        # то кол-во кобинаций = кол-ву комбинаций НЕбелых шаров в первой коробке
        if first_white_balls == 0 and second_white_balls > 0:
            first_var = np.math.comb(first_box_black,2-first_white_balls)
            second_var = np.math.comb(second_box_white,second_white_balls) * np.math.comb(second_box_black,4-second_white_balls)
        # если так случилось, что существует вариант, при котором из второй коробки может быть вытянуто 0 белых шаров,
        # то кол-во кобинаций = кол-ву комбинаций НЕбелых шаров во второй коробке
        if first_white_balls > 0 and second_white_balls == 0:
            second_var = np.math.comb(second_box_black,4-second_white_balls)
            first_var = np.math.comb(first_box_white,first_white_balls) * np.math.comb(first_box_black,2-first_white_balls)
        # иначе, кол-во вариантов = кол-ву вариантов вытягивания белых шаров из каждой коробки (без учета перестановок белых шаров),
        # умноженному на кол-во вариантов вытягивания НЕбелых шаров (без учета их перестановок) из этой же коробки
        if first_white_balls > 0 and second_white_balls > 0:
            first_var = np.math.comb(first_box_white,first_white_balls) * np.math.comb(first_box_black,2-first_white_balls)
            second_var = np.math.comb(second_box_white,second_white_balls) * np.math.comb(second_box_black,4-second_white_balls)
        print('Всего комбинаций для %d белых шаров из 1й коробки и %d белых шаров из 2й коробки:'%(first_white_balls, second_white_balls),second_var * first_var)
        # осталось только перемножить вычисленные кол-ва вариантов - и сложить все произведения на всех полученных итерациях, которые удовлетворяют условию
        total_var += second_var * first_var
'''
например, Б1,Б2,Ч1 (белые - Б, черные - Ч для первой коробки); ББ1, ББ2,ЧЧ1 (для второй коробки);
тогда кол-во вариантов вытянуть 2 белых (в сумме, ровно 2) из обоих, если вытягиваем 3 (2 из первой, 1 из второй):
Б1,Б2,ЧЧ1 (1 вариант), Б1,Ч1,ББ1 (2 вариант), Б2,Ч1,ББ1 (3 вариант), Б1,Ч1,ББ2 (4 вариант), Б2,Ч1,ББ2 (5 вариант) - всего 5 вариантов:
из них 4 - это когда 1 из шаров взят из 2й коробки, что получается перемножением вариантов: 2 (белых шара по-одному из одной коробки) * 2 (белых шара по-одному из второй коробки) = 4
и еще один вариант - когда оба шара достаются из первой коробки: 1 (2 белых шара из первой коробки) * 1 (вариант НЕбелый шар из второй коробки)

точно так же можно рассчитать кол-во вариантов вытянуть ровно 3 белых шара из обоих коробок: 1 (вариант вытянуть сразу 2 белых из первой коробки)
'''

print('Общее кол-во комбинаций для %d-%d шаров из двух коробок ='%(white_balls,numbers), np.math.comb(second_box,4) * np.math.comb(first_box,2))

print('\nВероятность вытянуть хотя бы 3 белых шара = ',total_var/ (np.math.comb(second_box,4) * np.math.comb(first_box,2)))
```

## 3

```python
'''
3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень.
Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6.
Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.
'''

import numpy as np
first = 0.9
second = 0.8
third = 0.6
# веростности попадания для каждого спортсмена сохраним в список sportsmen
sportsmen = np.array([first,second,third])
# длина списка определяется количеством спортсменов; вероятность того, что (любой) выстрел произвоел кто-то из них = 1/кол-во
# вероятность удачного выстрела:
success = np.array([s/len(sportsmen) for s in sportsmen]) # это вероятность того, что если стрелял кто-то из этих спортсменов, то он попадет
# вероятность неудачного выстрела
fall = np.array([(1-s)/len(sportsmen) for s in sportsmen]) # это вероятность того, что если стрелял кто-то из этих спортсменов, то он не попадет
# по условию задачи, выстрел произведен: (B = 1)
# по условию задачи, выстрел признан удачным (A= сумма sucsess: реализовались возможности попадания для всех спортсменов; мы не знаем какого)
B = 1
A = sum(success) # общая вероятность того, что кто-то из них попадет
P_A_B = success * B # это вероятность того, что при условии, что стрелял какой-то из трех спортсменов, он попал;
# в общем-то, это должно быть равно вероятности попадания для каждого спортсмена с учетом кол-ва спортсменов,
# которое определяет вероятность того, что он вообще стрелял

# следующим действием - рассчитывается вероятность стрельбы при попадании (формула Байеса)
P_B_A = B * P_A_B / A
print('Вероятность попадания: \n%.4f %s,\n%.4f %s,\n%.4f %s'%(P_B_A[0],'первый спортсмен',P_B_A[1],'второй спортсмен',P_B_A[2],'третий спортсмен.'))
```

```python
'''
4. В университет на факультеты A и B поступило равное количество студентов,
а на факультет C студентов поступило столько же, сколько на A и B вместе.
Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8.
Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9.
Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?
'''

import numpy as np
# веростности попадания для студентов каждого факультета сохраним в список facs
# студентов на первых двух факультетах поровну, а на третьем - вдвое больше
fac_a = 0.8 * 0.25
fac_b = 0.7 * 0.25
fac_c = 0.9 * 0.5
facs = np.array([fac_a,fac_b,fac_c]) # это же - вероятность того, что студент сдал сессию

# по условию задачи, сессию студент сдавал: (B = 1)
# по условию задачи, сдача сессии была успешной (реализовались все вероятности студентов всех факультетов)
A = sum(facs) # общая вероятность того, что студент какого-то факультета сдаст сессию
P_A_B = facs * 1 # это вероятность того, что при условии, что кто-то сдавал сессию - и сдал
# следующим действием - рассчитывается вероятность сдачи студентом определенного факультета при успешной сдаче
P_B_A = B * P_A_B / A
print('Рассчитанные вероятности: \n%.4f %s,\n%.4f %s,\n%.4f %s'%(P_B_A[0],'сдавал студент факультета A',P_B_A[1],'сдавал студент факультета B',P_B_A[2],'сдавал студент факультета C.'))
```
