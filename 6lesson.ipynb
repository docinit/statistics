## 1.

```python
import numpy as np
from scipy import stats as s

zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110],)
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])

covar_coef = ((zp - zp.mean()) * (ks - ks.mean())).sum() / (len(zp)-2)
pearson_coef = ((zp - zp.mean()) * (ks - ks.mean())).sum() / (((zp - zp.mean())**2).sum() * ((ks - ks.mean())**2).sum())**0.5
print(f'''
1. Коэффициент ковариации:
Без использования библиотеки numpy: {covar_coef}
С помощью numpy: {np.cov(np.array([zp,ks]),ddof=2)[0,1]}

###########

2. Коэффициент корреляции Присона:
Без использования библиотеки numpy: {pearson_coef}
С помощью numpy: {np.corrcoef([zp,ks])[0,1]}
''')
```

## 2.

```python
iqs = np.array([131, 125, 115, 122, 131, 115, 107, 99, 125, 111])
mean = sum(iqs)/len(iqs)
std = (sum([(i - mean)**2 for i in iqs])/ (len(iqs)-1)) **0.5
mse = std / (len(iqs)**0.5)

z = s.t.ppf(0.975, df=len(iqs)-1)
print("ДИ = (",mean-z*mse,";",mean+z*mse,")")
```
